```python
import pandas as pd

df = pd.read_csv('olista_data.csv')
df.head()
```

```python
df['delivery_status'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['delivery_status'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Estado de Entregas')
plt.show()
```

```python
late_df = df[df['delivery_status'] == 'late']
late_df['purchase_dayofweek'].value_counts().plot(kind='bar', color='tomato')
plt.title('Pedidos Tarde por Día de la Semana')
plt.xlabel('Día')
plt.ylabel('Cantidad')
plt.tight_layout()
plt.show()
```

```python
late_df['purchase_hour'].value_counts().sort_index().plot(kind='line', color='tomato')
plt.title('Pedidos Tarde por Hora del Día')
plt.xlabel('Hora')
plt.ylabel('Cantidad')
plt.tight_layout()
plt.show()
```
